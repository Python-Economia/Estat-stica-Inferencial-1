```python
'''
Quocientes de Inteligência (QI's) medidos no Stanford Revision of the Binet-Simon Intelligence Scale são normalmente 
distribuídos com média 100 e desvio padrão 16.
Determine a porcentagem de pessoas que possuem QI's entre 115 e 140.
'''
```

### Resposta sugerida

```python
import scipy.stats as stats
```

```python
'''
A variável QI é normalmente distribuída (com média 100 e desvio padrão 16).
Para uma variável aleatória contínua, a área sob o gráfico da sua função densidade de probabilidade num dado intervalo nos dá a
porcentagem que aquele intervalo representa em relação à todos os possíveis valores que a variável pode assumir. 
Como os QI's são normalmente distribuídos com média 100 e desvio padrão 16, a porcentagem de pessoas que possuem QI's entre 115 
e 140 é dada pela área sob a curva de uma distribuição normal com média 100 e desvio padrão 16 no intervalo entre 115 e 140.
Esta área pode ser vista no gráfico (região listrada). O cálculo é feito abaixo.
'''
```

```python
# Cálculo da região listrada no gráfico (intervalo entre 115 e 140)
area_abaixo_140 = stats.norm.cdf(x=140, loc=100, scale=16)  # Calcula a área abaixo de 140
area_abaixo_115 = stats.norm.cdf(x=115, loc=100, scale=16)  # Calcula a área abaixo de 115  
probabilidade = area_abaixo_140 - area_abaixo_115           # Calcula a área abaixo de 140 retirando-se a a área abaixo de 115  
probabilidade
```

```python
'''
Conclusão
A área listrada no gráfico (entre 115 e 140) é de 0.1680.
Assim, concluímos que 16,80% de todas as pessoas possuem QI's entre 115 e 140.

Também podemos interpretar essa área da seguinte forma:
A probabilidade é de 16,80% que uma pessoa selecionada aleatoriamente tenha um QI entre 115 e 140.
'''
```
